## Loading data in pandas


```python
import pandas as pd
test = pd.read_csv("Test.csv")
test
```

# Missing Value

```python
test.isna().sum()
```

```python
#We can see that Item_Weight and Outlet_Size have missing values. We can fill them using mean and mode
```

```python
test['Item_Weight'] = test['Item_Weight'].fillna(test['Item_Weight'].mean())
  
# replacing missing values in price column
# with median of that column
test.fillna(test['Outlet_Size'].mode()[0], inplace=True)
```

```python
#Now, no missing values.
test.isna().sum()
```

# Feature Preprocessing, Encoding, Scaling

```python
test
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
test['Outlet_Establishment_Year'] = le.fit_transform(test['Outlet_Establishment_Year'])
```

```python
test['Item_Fat_Content'] = le.fit_transform(test['Item_Fat_Content'])
```

```python
test['Outlet_Size'] = le.fit_transform(test['Outlet_Size'])
```

```python
test['Outlet_Location_Type'] = le.fit_transform(test['Outlet_Location_Type'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
test['Item_Weight'] = scaler.fit_transform(test[['Item_Weight']])
test['Item_MRP'] = scaler.fit_transform(test[['Item_MRP']])
```

```python
import category_encoders as ce

encoder = ce.BinaryEncoder(cols=['Item_Type'])
test = encoder.fit_transform(test)
```

```python
encoder = ce.BinaryEncoder(cols=['Outlet_Type'])
test = encoder.fit_transform(test)

encoder = ce.BinaryEncoder(cols=['Outlet_Identifier'])
test = encoder.fit_transform(test)
```

```python
test.drop(columns = ['Item_Identifier'], axis = 1, inplace = True)
```
