## 5).Exploratory Data Analysis

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Plot histograms or density plots
train[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year', 'Item_Outlet_Sales']].hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# Plot box plots
plt.figure(figsize=(10, 8))
sns.boxplot(data=train[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year', 'Item_Outlet_Sales']])
plt.tight_layout()
plt.show()

#A z score is a measure of a value and how far it is from a mean. The standard deviation is normalized. A z score with a value
#more than 3 is 3 STDVs away from the mean. That can be considered an outlier.

# Calculate z-scores and flag potential outliers
num_cols = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year', 'Item_Outlet_Sales']
for col in num_cols:
    z = np.abs((train[col] - train[col].mean()) / train[col].std())
    train[col+'_zscore'] = z
    outliers = train[z > 3]
    print(f'{col}: {len(outliers)} potential outliers')
```

# Missing Value

```python
#We can see that Item_Weight and Outlet_Size have missing values. We can fill them using mean and mode
```

```python
train['Item_Weight'] = train['Item_Weight'].fillna(train['Item_Weight'].mean())
  
# replacing missing values in price column
# with median of that column
train.fillna(train['Outlet_Size'].mode()[0], inplace=True)
```

```python
import category_encoders as ce

encoder = ce.BinaryEncoder(cols=['Item_Type'])
train = encoder.fit_transform(train)
```

```python
encoder = ce.BinaryEncoder(cols=['Outlet_Type'])
train = encoder.fit_transform(train)

encoder = ce.BinaryEncoder(cols=['Outlet_Identifier'])
train = encoder.fit_transform(train)
```

```python
train.drop(columns = ['Item_Identifier'], axis = 1, inplace = True)
```

# Training some models on this dataframe and saving them for testing

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from xgboost import XGBRegressor
```

```python
X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print('Linear Regression Training R^2:', lr.score(X_train, y_train))

# Fit a regularized linear regression model
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
print('Lasso Training R^2:', lasso.score(X_train, y_train))

# Fit a random forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
print('Random Forest Training R^2:', rf.score(X_train, y_train))

#Fit an xgb Regressor model
xgb_model = xgb.XGBRegressor()
xgb_model.fit(X_train, y_train)
print('XGBoost R^2:', xgb_model.score(X_train, y_train))
```

```python
import pickle

# save linear regression model
with open('linear_regression.pkl', 'wb') as file:
    pickle.dump(lr, file)
    
# save regularized linear regression model
with open('lasso.pkl', 'wb') as file:
    pickle.dump(lasso, file)
    
# save random forest model
with open('random_forest.pkl', 'wb') as file:
    pickle.dump(rf, file)
    
# save xgboost model
with open('xgboost.pkl', 'wb') as file:
    pickle.dump(xgb_model, file)
```
