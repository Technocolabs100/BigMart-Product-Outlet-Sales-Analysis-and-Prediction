## Remove Outliers

### Item Weight

```python
df.boxplot(column=['Item_Weight'])
```

```python
def remove_outlier(col):
  sorted(col)
  Q1,Q3=col.quantile([0.25,0.75])
  IQR=Q3-Q1
  lr=Q1-(1.5*IQR)
  ur=Q3+(1.5*IQR)
  return lr,ur
```

```python
import numpy as np
lw,uw=remove_outlier(df['Item_Weight'])
df['Item_Weight']=np.where(df['Item_Weight']>uw,uw,df['Item_Weight'])
df['Item_Weight']=np.where(df['Item_Weight']<lw,lw,df['Item_Weight'])
```

```python
df.boxplot(column=['Item_Weight'])
```

### Item Visibility

```python
df.boxplot(column=['Item_Visibility'])
```

# Modelling

## Linear Regeression

### Normal

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = df.loc[:, df.columns != 'Item_Outlet_Sales']
```

```python
x.head()
```

```python
y=df['Item_Outlet_Sales']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train, Y_test= train_test_split(x,y,random_state= 101,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train_std= sc.fit_transform(X_train)
X_test_std= sc.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_std, Y_train)
model = LinearRegression().fit(X_train_std, Y_train)
```

```python
r_sq = model.score(X_train_std, Y_train)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
y_pred = model.predict(X_test_std)
r2_score(Y_test,y_pred)
```

```python
mean_absolute_error(Y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(Y_test,y_pred))
```

### Using SM

```python
import statsmodels.api as sm
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
print(results.summary())
```

```python
print(f"coefficient of determination: {results.rsquared}")
print(f"adjusted coefficient of determination: {results.rsquared_adj}")
print(f"regression coefficients: {results.params}")
```

```python
print(f"predicted response:\n{results.predict(x)}")
```

## Random Forrest

```python
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor()
RF.fit(X_train,Y_train)
Y_pred_RF=RF.predict(X_test)
r2_score(Y_test,Y_pred_RF)
```

```python
mean_absolute_error(Y_test,Y_pred_RF)
np.sqrt(mean_squared_error(Y_test,Y_pred_RF))
```
